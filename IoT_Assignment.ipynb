## Task 1: Calculate Seconds in Given Days

```python
%%writefile main.cpp
#include <iostream>
using namespace std;
int main() {
    int days;
    cout << "Enter the number of days: ";
    cin >> days;
    int seconds = days * 24 * 60 * 60;
    cout << "Number of seconds in " << days << " days is: " << seconds << endl;
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

## Task 2: Calculate Volume of a Sphere


```python
%%writefile main.cpp
#include <iostream>
#include <cmath>
using namespace std;
int main() {
    double radius;
    cout << "Enter the radius of the sphere: ";
    cin >> radius;
    double volume = (4.0 / 3.0) * M_PI * pow(radius, 3);
    cout << "Volume of the sphere is: " << volume << endl;
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

## Task 3: Area and Perimeter of a Square

```python
%%writefile main.cpp
#include <iostream>
using namespace std;
int area(int side) { return side * side; }
int perimeter(int side) { return 4 * side; }
int main() {
    int side;
    cout << "Enter the side length of the square: ";
    cin >> side;
    cout << "Area: " << area(side) << endl;
    cout << "Perimeter: " << perimeter(side) << endl;
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

## Task 4: Determine Character Case

```python
%%writefile main.cpp
#include <iostream>
using namespace std;
void checkCase(char c) {
    if (c >= 'A' && c <= 'Z') {
        cout << c << " is uppercase." << endl;
    } else if (c >= 'a' && c <= 'z') {
        cout << c << " is lowercase." << endl;
    } else {
        cout << c << " is not an alphabetic character." << endl;
    }
}
int main() {
    char ch;
    cout << "Enter a character: ";
    cin >> ch;
    checkCase(ch);
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

## Task 5: Pseudocode to C++

```python
%%writefile main.cpp
#include <iostream>
using namespace std;
int main() {
    double x = 0;
    double y = 20;
    do {
        y = y - 4;
        x = x + 2 / y;
    } while (y >= 6);
    cout << "x = " << x << endl;
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

## Task 6: Array Input and Average Calculation

```python
%%writefile main.cpp
#include <iostream>
using namespace std;
int main() {
    const int SIZE = 5;
    double arr[SIZE];
    double sum = 0;
    cout << "Enter 5 values:" << endl;
    for (int i = 0; i < SIZE; ++i) {
        cin >> arr[i];
        sum += arr[i];
    }
    double average = sum / SIZE;
    cout << "Average = " << average << endl;
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```
